# Simpler Practice of List

## 1. create a list
Give a integer n, to create a list which consist of positive integer squares of length n

e.g. n=3, r = [1, 4, 9]


```python
n = 3
r = []
for i in range(1,n+1):
    r.append(i**2)

print(r)
```

```python
n=6
r = [ i**2 for i in range(1,n+1)]
print(r)
```

## 2. modify a list
change all the even number in above list to str type

e.g. r = [1, '4', 9]

```python
r
```

```python
i=0
while i<len(r):
    if r[i]%2 == 0:
        # print(r[i])
        s = str(r[i])
        r[i] = s
    i = i + 1
```

```python
for i in r:
    print(i)
```

### then we can draw graph with list!

```python
size = 20
```

```python
a1 = plt.imread('1.png')
```

```python
a1.shape
```

```python
a1
```

```python
a = [ [0 for i in range(size)] for i in range(size)]
```

```python
x = []
y = []

for i in range(size):
    for j in range(size):
        if sum(a1[i][j])!=3:
            a[i][j]=1
            y.append(size-i)
            x.append(j)
```

```python
a
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y,vmin=0, vmax=100)

ax.set(xlim=(0, size),
       ylim=(0, size))

plt.show()
```
